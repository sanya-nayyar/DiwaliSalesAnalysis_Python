**Exploratory Data Analysis**

**Gender**

```python
df.columns
```

```python
ax=sns.countplot(x='Gender', data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].mean().sort_values(by='Amount',ascending=False)

sns.barplot(x='Gender',y='Amount',data=sales_gen)
```

_From above graphs we can see that majority of buyers are female and even the purchasing power of females is greater than that of men_

### Age

```python
ax=sns.countplot(data=df,x='Age Group',hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby('Age Group', as_index=False)['Amount'].mean()
```

On an average, people from 55+ age category makes an order of higher amount as compared to other age categories.

```python
df.groupby('Gender', as_index=False)['Amount'].mean()
```

On an average, female makes a higher amount of order than men by 124 rupees.

```python
#Total Amount Vs Age
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group', y= 'Amount', data=sales_age)
```

```python
axx=df.groupby(['Age Group','Gender'],as_index=False)['Amount'].mean().sort_values(by='Amount',ascending=False)
axx
sns.barplot(data=axx,x='Age Group', y='Amount',hue='Gender')
```

From above graphs highest amount orders are made by men from the age category of 55+ whereas female's highest order is from the age group 36-45 yrs.

### States

```python
df.columns
```

```python
sales_state=df.groupby(['State'],as_index=False)['Orders'].mean().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=sales_state,x='State',y='Orders')
```

```python
sales_state=df.groupby(['State'],as_index=False)['Amount'].mean().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=sales_state,x='State',y='Amount')
```

Orders are approximately equally distributed among top 10 states
Delhi has higher amount of order generally.

```python
ax= sns.countplot(data=df,x='Marital_Status')

sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

### Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})
ax=sns.countplot(data=df,x='Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state=df.groupby(['Occupation'],as_index=False)['Amount'].mean().sort_values(by='Amount',ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_state,x='Occupation',y='Amount')
```

On an average, people with occupation as govt has orders with higher amount than people with other occupations

###  Product Category

```python
sns.set(rc={'figure.figsize':(20,5)})
ax=sns.countplot(data=df,x='Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state=df.groupby(['Product_Category'],as_index=False)['Amount'].mean().sort_values(by='Amount',ascending=False)

sns.set(rc={'figure.figsize':(25,5)})
sns.barplot(data=sales_state,x='Product_Category',y='Amount')
```

On an average,products are from auto, tupperware and Hand&Power tools

```python
sales_state=df.groupby(['Product_ID'],as_index=False)['Amount'].mean().sort_values(by='Amount',ascending=False).head(15)

sns.set(rc={'figure.figsize':(25,5)})
sns.barplot(data=sales_state,x='Product_ID',y='Amount')
```
